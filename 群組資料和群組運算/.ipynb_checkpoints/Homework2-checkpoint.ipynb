```python
import numpy as np
import pandas as pd

#美國球賽門票收入
tips = pd.read_csv('tips.csv')
tips
```

```python
#改變欄位
tips.columns = ['總票價', '小費', '吸煙者', '日期', '時間', '大小']
tips
```

```python
#增加小費但總費用的比例
tips['小費比例'] = tips['小費'] / tips['總票價']
tips
```

```python
#以日期和吸煙者做為group的key,建立Groupby物件
grouped = tips.groupby(['日期', '吸煙者'])

#取出grouped內的小費欄位
grouped_pct = grouped['小費']

#使用agg(),傳入內建的function mean
#求出平均小費
grouped_pct.agg('mean')
```

```python
#使用agg一次傳入多個function
#使用list
def peak_to_peak(arr):    
    return arr.max() - arr.min()

#標準差是一組數值自平均值分散開來的程度的一種測量觀念。一個較大的標準差，代表大部分的數值和其平均值之間差異較大；一個較小的標準差，代表這些數值較接近平均值。
#兩組數的集合 {0, 5, 9, 14} 和 {5, 6, 8, 9} 其平均值都是 7 ，但第二個集合具有較小的標準差

grouped_pct.agg(['mean', 'std', peak_to_peak])
```

```python
#讓產生的欄位有自訂的欄位名稱
#list內使用2-tuples
grouped_pct.agg([('平均','mean'),('標準差',np.std)])
```

```python
#一個使用2個欄位和多個自訂的function
functions = [('數量','count'), ('平均','mean'), ('最大值','max')]
result = grouped['小費','總票價'].agg(functions)
result
```

```python
#單獨取出小費
result['小費']
```

```python
#計算各數值與平均數的差，取其平方後加總，再除以數值個數，得「變異數」。變異數開根號後得「標準差」。
ftuples = [('平均', 'mean'), ('變異數',np.var), ('標準差', np.std)]
grouped['小費','總票價'].agg(ftuples)
```
