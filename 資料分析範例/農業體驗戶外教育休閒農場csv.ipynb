```python
%matplotlib notebook
import requests
import pandas as pd
import csv
import matplotlib.pyplot as plt
url = 'http://data.coa.gov.tw/Service/OpenData/ODwsv/ODwsvOutdoorEdu.aspx?FOTT=CSV'
r=requests.get(url,verify=False)
decoded_content =r.content.decode('utf-8')
cr = csv.reader(decoded_content.splitlines(), delimiter=',')
with open('output1.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(cr)
df = pd.read_csv('output1.csv', sep=',',index_col='OutdoorEduID')
df.sort_index()
```

```python
df1=df.reindex(columns = ['FarmNm_CH','County'])
df2=df1['County'].value_counts()
figure1, ax1 = plt.subplots()
df2.plot(ax=ax1, kind='bar')
```
