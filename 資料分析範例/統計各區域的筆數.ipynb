```python
#分析文字檔內容,顯示第一行內容
path = 'datasets/bitly_usagov/example.txt'
with open(path) as file:
    print(file.readline())
```

```python
#將第一行的文字轉為Dictionary,並且放置到list內
#產生list內有Dictionary

import json
path = 'datasets/bitly_usagov/example.txt'
records = [json.loads(line) for line in open(path)]
records[0]
```

```python
#目標收集dictionary的'tz'的value,成為list
#使用list comprehension過濾出資料

time_zones = [rec['tz'] for rec in records] #出現錯誤,有些Dictionary沒有key:'tz'
```

```python
#加上檢查是否Dictionary有key:tz
time_zones = [rec['tz'] for rec in records if 'tz' in rec]
time_zones[:10]
```

```python
import pandas as pd #使用pandas取出前10名
frame = pd.DataFrame(records) #pandas的Structure
frame.info()#使用info()方法查看index,查看筆數3560,查看columns的label名和每個columns資料的dtype和數量
```

```python
#取出key:'tz'前10筆
frame['tz'][:10]
```

```python
#取出各區域的筆數
tz_counts = frame['tz'].value_counts() #value_counts計算內容相同的數量,產生新的Series,原來的值成為Series的index,value成為數量
tz_counts
```

```python
#將內容是None的填入'遺失的資料'
clean_tz = frame['tz'].fillna('遺失的資料')

#將空字串的區域填入'不明'
clean_tz[clean_tz == ''] = '不明'
clean_tz[:20]
```

```python
tz_counts = clean_tz.value_counts()
tz_counts
```

```python
#只分析前10筆
tz_counts[:10]
```
