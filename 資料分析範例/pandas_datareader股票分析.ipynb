```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader.data as web

all_data = {ticker:web.get_data_yahoo(ticker,start='2019/1/1') for ticker in ['1301.TW','1303.TW','1304.TW','1305.TW']}
price = pd.DataFrame({ticker:data['Adj Close']for ticker,data in all_data.items()})
volume = pd.DataFrame({ticker:data['Volume'] for ticker,data in all_data.items()})
price.columns = ['台塑','南亞','台聚','華夏']
volume.columns = ['台塑','南亞','台聚','華夏'] 
```

```python
price[-10:].plot()
```

```python
volume.tail().plot()
```

```python
priceDiff = price.diff()
```

```python
volumeDiff = volume.diff()
```

```python
priceDiffPercent = priceDiff / price
priceDiffPercent
```

```python
volumeDiffPercent = volumeDiff / volume
volumeDiffPercent
```

```python
priceDiffPercent.tail().plot()
```
