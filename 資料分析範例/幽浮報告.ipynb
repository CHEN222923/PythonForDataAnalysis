```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("http://bit.ly/uforeports")
df1 = df.set_index(['State', 'City'])
df1
```

```python
df1.sort_index(level=0)
```

```python
df2 = df1.groupby(level=['State']).count()
df3 = df2['Time']
```

```python
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=15)
figure = plt.figure()
ax1 = figure.add_subplot(1, 1, 1)
df3.plot.barh(ax = ax1)
ax1.figure.set_size_inches(25, 40)
ax1.set_title("美國各州幽浮出現次數",fontproperties=font)
df4 = df1.groupby(level=['State','City']).count()
figure1 = plt.figure()
figure1.set_size_inches(25, 40)
states = np.sort(df['State'].unique())
ax2 = figure1.add_subplot(1,1,1)
ax2.figure.set_size_inches(25, 40)
ax2.set_title("AK各城幽浮出現次數",fontproperties=font)
stateFrame = df4.loc['AK','Time'] 
stateFrame.plot.barh(ax = ax2)

```
