```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
```

```python
xlsx = pd.ExcelFile('總統.xls')
all = pd.read_excel(xlsx, '中　央')
all
```

```python
df = all.iloc[5:,:4]
df.columns = ['縣市','朱立倫','蔡英文','宋楚瑜']
df.name = '第14任總統、副總統選舉候選人得票數一覽表'
df
```

```python
df=df.reset_index().drop('index',axis=1)
df
```

```python
df = df.set_index('縣市')
df.columns.name = '總統候選人'
df
```

```python
df2=df.sum()
```

```python
figure2, ax2 = plt.subplots()
df2.plot(ax = ax2,kind='bar',color=['b','g','orange'])
ax2.set_xticklabels(df2.index,fontproperties=font,rotation=30)
ax2.set_xlabel('總統候選人',fontproperties=font)
for p in ax2.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax2.annotate('{:}票'.format(height), (x+width/10, height/2),fontproperties=font)
```
