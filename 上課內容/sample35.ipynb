```python
#空氣品質指標(AQI)
import csv
import pandas as pd
import urllib3
import certifi

CSV_URL = "https://opendata.epa.gov.tw/api/v1/AQI?%24skip=0&%24top=1000&%24format=csv"
http = urllib3.PoolManager(cert_reqs='CERT_REQUIRED',ca_certs=certifi.where())
r = http.request('GET', CSV_URL)
if r.status == 200:
    with open('output.csv','wb') as f:
        f.write(r.data)
        f.close()
else:
    print('下載錯誤')
    
df = pd.read_csv('output.csv', sep=',')
df.head(20)
```

```python
df.info()
```

```python
final_df = df.loc[:,['County', 'SiteName', 'AQI', 'PublishTime', 'PM2.5_AVG']]
final_df.columns = ['縣市', '位置', '空氣品質', '發佈時間', 'PM2.5']
final_df
```

```python
def f(x):
    if x <= 50:
        return '良好'
    elif x <=100:
        return '普通'
    elif x <= 150:
        return '對敏感族群不健康'
    elif x <= 200:
        return '對所有族群不健康'
    elif x <= 300:
        return '非常不健康'
    else: 
        return '危害'
        
final_df['空氣狀況'] = final_df['空氣品質'].map(f)
final_df.head(20)
```

```python
final_df.index = final_df['縣市'] + final_df['位置']
final_df
```

```python
df1=final_df.drop(['縣市', '位置'],axis=1)
df1
```

```python
stateFrame = df1.groupby('空氣狀況').size()
stateFrame
```

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=14)

figure, ax1 = plt.subplots()
stateFrame.plot(ax=ax1,kind='bar',color=['r','g','orange', 'blue'])
#ax1.legend(prop=font)
ax1.set_title('空氣狀況',fontproperties=font)
ax1.set_xticklabels(stateFrame.index, fontproperties=font,rotation=0)
ax1.set_xlabel('狀況', fontproperties=font)
```

```python
for p in ax1.patches:
    width, height = p.get_width(), p.get_height()
    #print('width',width)
    #print('height',height)
    x, y = p.get_xy()
    #print('x',x)
    #print('y',y)
    #print('x+width/2',x+width/2)
    #print('y+height', x+height)
    ax1.annotate('{:}'.format(height), (x+width/2, y+height + height/50), fontproperties=font1)
```
