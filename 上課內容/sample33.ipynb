```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

index = [('台北', 2019), ('台北', 2020),
         ('台中', 2019), ('台中', 2020),
         ('台南', 2019), ('台南', 2020)]
populations = [33871648, 37253956,
               18976457, 19378102,
               20851820, 25145561]
index = pd.MultiIndex.from_tuples(index,names = ['縣市', '年'])
pop = pd.Series(populations, index=index)
pop
```

```python
pop.unstack() #運算式
pop_df = pop.unstack() #敘述式
pop_df
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
pop_df.plot()
```
