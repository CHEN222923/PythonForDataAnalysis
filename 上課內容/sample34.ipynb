```python
#先前整理資料

def f(x):
    if ',' in x:
        l = x.split(',')
        return int(''.join(l))
    else:
        return int(x)

df['週六'] = df['週六'].map(f)
df['週日'] = df['週日'].map(f)
df.info()
```
